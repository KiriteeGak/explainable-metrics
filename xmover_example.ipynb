```python
from xmover_explainer import ExplainableXMover

model = ExplainableXMover('de','en')
src = 'Er mag Hunde' # he likes dogs
trans = 'He dislikes dogs'
score = model(src, trans)
print('xmover score:', score)

exps = model.explain(src, trans, plot=True)
```

```python
trans = 'He loves dogs'
score = model(src, trans)
print('xmover score:', score)

exps = model.explain(src, trans, plot=True)
```
